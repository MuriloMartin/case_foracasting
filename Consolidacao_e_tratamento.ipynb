# Variaveis explicativas

```python
def standardize_time_format(time_str):
    if len(time_str) == 5:  # Format HH:mm
        return time_str
    else:   # Format HHmm
        return time_str[:2] + ':' + time_str[2:4]
def standardize_date_format(date_str):
    return date_str.replace('/','-')

def get_season(month):
    if month in [12, 1, 2]:
        return 1
    elif month in [3, 4, 5]:
        return 2
    elif month in [6, 7, 8]:
        return 3
    elif month in [9, 10, 11]:
        return 4
        
```

```python
dfs_exp = []
encoding = 'ISO-8859-1'
dfs_exp.append(pd.read_csv('C:\\Users\\Murilo\\Desktop\\Projetos\\DesafioPSR\\datasets\\exp_pirapora-2018.csv', sep=';', encoding = encoding, skiprows=8, index_col=None))
dfs_exp.append(pd.read_csv('C:\\Users\\Murilo\\Desktop\\Projetos\\DesafioPSR\\datasets\\exp_pirapora-2019.csv', sep=';', encoding = encoding, skiprows=8, index_col=None))
dfs_exp.append(pd.read_csv('C:\\Users\\Murilo\\Desktop\\Projetos\\DesafioPSR\\datasets\\exp_pirapora-2020.csv', sep=';', encoding = encoding, skiprows=8, index_col=None))
dfs_exp.append(pd.read_csv('C:\\Users\\Murilo\\Desktop\\Projetos\\DesafioPSR\\datasets\\exp_pirapora-2021.csv', sep=';', encoding = 'utf-8', skiprows=8, index_col=None))
```

```python
for i in range(len(dfs_exp)):
    dfs_exp[i].columns = ['Data',
'hora',
'Precipitacao',
'Pressao ATM (mB)',
'Pressao ATM MAX.NA HORA ANT. (AUT) (mB)',
'Pressao ATM MIN NA HORA ANT. (AUT) (mB)',
'RADIACAO GLOBAL (Kj/m2)',
'TEMPERATURA DO AR - BULBO SECO, HORARIA (C)',
'TEMPERATURA DO PONTO DE ORVALHO (C)',
'TEMPERATURA MAX NA HORA ANT. (AUT) (C)',
'TEMPERATURA MIN NA HORA ANT. (AUT) (C)',
'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (C)',
'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (C)',
'UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)',
'UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)',
'UMIDADE RELATIVA DO AR, HORARIA (%)',
'VENTO, DIRECAOO HORARIA',
'VENTO, RAJADA MAXIMA (m/s)',
'VENTO, VELOCIDADE HORARIA (m/s)']
    dfs_exp[i]['hora'] = dfs_exp[i]['hora'].apply(standardize_time_format)
    dfs_exp[i]['Data'] = dfs_exp[i]['Data'].apply(standardize_date_format)
```

```python
df_exp = pd.concat(dfs_exp)
df_exp['DATETIME'] = pd.to_datetime(df_exp['Data'] + ' ' + df_exp['hora'])
# Set the new datetime column as the index
df_exp.set_index('DATETIME', inplace=True)
# Shift the datetime index by 3 hours earlier
df_exp.index = df_exp.index - pd.Timedelta(hours=3)
```

```python
df_exp['hora'] = df_exp.index.hour
df_exp['dia'] = df_exp.index.day
df_exp['mes'] = df_exp.index.month
df_exp['ano'] = df_exp.index.year
```
